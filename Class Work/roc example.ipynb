# ROC Curve in class example

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
# y is classes, 1 for + class (F), 0 for - class (M)
y = np.array([1,1,0,1,1,0,0,0,1,0])
```

```python
# scores for each point 
y_score = np.array([0.98,0.92,0.83,0.83,0.6,0.49,0.38,0.26,0.26,0.26])
```

```python
fpr, tpr, thresh = metrics.roc_curve(y,y_score)
```

```python
roc_auc = metrics.auc(fpr,tpr)
print(roc_auc)
```

```python
plt.figure(figsize=(6,6))
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.3f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")

for i in range(0, len(thresh),1 ):
    plt.annotate('%0.2f' % thresh[i],xy=(fpr[i]-0.00,tpr[i]-0.03),
                 textcoords='data',fontsize=12,color='red', rotation=-45  )
plt.show()
```

The ROC curve shows that a threshold of 0.6 gives an optimal accuracy.  And we see the effects of raising or lowering the threshold
