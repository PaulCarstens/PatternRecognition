# ROC Curves

Evaluating and tuning classifiers based on ROC curves

```python
# Simplified from Tutorial 3-1

from sklearn.datasets import make_moons
from sklearn import svm
from sklearn import metrics

X, y = make_moons(noise=0.5, random_state=0)
clf = svm.SVC(kernel='linear')  
clf.fit(X,y)
y_score = clf.decision_function(X)
```

```python
y_score[:10]
```

```python
y_pred = clf.predict(X)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plot_colors = "ryb"
plot_step = 0.02

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(12,12))
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu)

plt.scatter(X[:, 0], X[:, 1], marker='o', c=y,
            s=25, edgecolor='k')

 #   plt.annotate('%3.1f' % y_score[i],xy=(X[0,i],X[1,i]),textcoords='data'  )

for i in range( len(y) ):
    plt.annotate('%0.1f' % y_score[i],xy=(X[i,0]+0.05,X[i,1]),textcoords='data'  )

plt.show()
```

I worked really hard on this, so let's stop and appreciate it for awhile.  We used the *decision_function* to compute a score for each point.  Points in the blue area are more positive, red area is negative, and zero near the boundary.  Not only do we use the score to classify each point, but we get a measure of how good the classification guess was!

```python
metrics.confusion_matrix(y,y_pred)
```

```python
print(metrics.classification_report(y, y_pred))
```

# ROC Curves

The boundary was set by minimizing total errors for the training set (treating both classes equally), and the score is just the distance to the boundary. We use the score to make decisions, and this gives us a certain TPR and FPR.  But we could make other choices...

```python
metrics.roc_curve?
```

```python
fpr, tpr, thresh = metrics.roc_curve(y,y_score)
```

```python
# Area Under Curve
roc_auc = metrics.auc(fpr,tpr)
print(roc_auc)
```

```python
plt.figure(figsize=(6,6))
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.3f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")

# annotate thresholds at every other point
for i in range(1, len(thresh),2 ):
    plt.annotate('%0.1f' % thresh[i],xy=(fpr[i]+0.01,tpr[i]-0.03),
                 textcoords='data',fontsize=10,color='red', rotation=-45  )

plt.show()
```

We can make sense of the result by carefully comparing the last two plots.

The choices for thresholds occur where lowering threshold accepts next point of a different class.

```python
# Create an array with 2 columns for scores and class, then sort it by score
scores = np.zeros((100,2))
scores[:,0] = y_score
scores[:,1] = y
s = scores[scores[:,0].argsort()] # found on Stack Exchange, of course
```

Another way to thkink about ROC curve:
* Make this sorted array
* Start drawing ROC curve at 0,0
* Whenever you get a point right move up
* Whenever you get a point wrong move right
